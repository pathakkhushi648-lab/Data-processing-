```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Income',
    y='PurchaseAmount',
    hue='CustomerSatisfaction',
    palette='viridis',
    s=100
)

plt.title('Income vs Purchase Amount by Customer Satisfaction')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='MembershipType',
    data=df,
    palette='viridis'
)

plt.title('Membership Type Distribution')

plt.show()
```

The boxplot indicates that customers with lower satisfaction scores are more likely to churn. Customer satisfaction appears to be an important factor influencing retention.

# Framing the Machine Learning Problem from Raw Data

## Problem Definition

The raw dataset contains customer demographic information, purchasing behavior, membership details, website activity, and customer satisfaction scores. The objective is to analyze this raw data and develop a machine learning model that can predict whether a customer is likely to churn.

## Business Problem

Customer churn is a major challenge for businesses because losing customers directly affects revenue and profitability. Identifying customers who are likely to leave allows organizations to take proactive measures such as personalized offers, loyalty programs, and customer engagement strategies.

## Machine Learning Problem

The problem is framed as a **Supervised Machine Learning Classification Problem** because the dataset contains historical customer records along with a target variable indicating whether the customer churned or not.

### Target Variable

**Churn**

* 0 = Customer Stays
* 1 = Customer Churns

### Input Features

* Age
* Gender
* Income
* MembershipType
* PurchaseAmount
* Quantity
* CustomerSatisfaction
* WebsiteVisitsPerMonth
* LastPurchaseDaysAgo
* EmailSubscribed
* PreferredPaymentMethod
* City
* State

## Input and Output

### Input (X)

Customer demographic and behavioral features.

### Output (Y)

Customer Churn Status

* Yes (1)
* No (0)

## Type of Machine Learning

Supervised Learning

## Type of Problem

Binary Classification

## Success Criteria

The success of the model will be evaluated using:

* Accuracy
* Precision
* Recall
* F1-Score
* Confusion Matrix

A model with high accuracy and strong precision/recall values will be considered effective for predicting customer churn.

## Expected Outcome

The final machine learning model should be able to predict whether a customer is likely to churn based on historical customer behavior and demographic information. This prediction can help businesses improve customer retention and reduce revenue loss.
